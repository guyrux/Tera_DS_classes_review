```python
# De posse do arquivo csv, a primeira coisa a se fazer é carregar o dataset na memória na forma de um dataframe.
df = pd.read_csv("data/gastos-deputados-20190406.csv")

# Para verificarmos se os dados foram corretamente carregados, podemos fazer rápidas verificações de como são os dados:
df.head(3).T
```

Na saída acima, podemos ter uma ideia da natureza dos dados brutos do dataset.
O tipo *int64* refere-se a números inteiros; *float64* refere-se a valores com ponto flutuante; e *object* é pode indicar tratar-se de uma string. Para matar a dúvida, sobre o conteúdo de cada atributo, pode-se plotar aquele atributo.

```python
df.datemissao.head(3)
# Podemos observar que apesar de tratar-se de uma data, o tipo desse atributo foi classificado como object.
# Nesse caso, é interessante transformarmos em data.
```

```python
# Convertendo datemissao para dias:
df.datemissao = df.datemissao.astype("datetime64[D]")
df.datemissao.head(3)
```

```python
df["DiaDaSemana"] = df.datemissao.dt.weekday
```

```python
df["DiaDaSemana"].replace([0,1,2,3,4,5,6],
                          ["Segunda-feira","Terça-feira","Quarta-feira","Quinta-feira","Sexta-feira",
                           "Sábado","Domingo"],inplace=True);
```

```python
df["NomeEPartido"] = df.txnomeparlamentar+" | "+df.sgpartido
```

```python
df.info()
```

```python
# O método DataFrame.describe(), descreve estatisticamente os atributos numéricos:
df.describe().T
```

```python
lista_MaioresGastadores = df.groupby("NomeEPartido").sum().sort_values(by="vlrliquido",
                                                                       ascending=False)["vlrliquido"].head(10)
lista_Nomes = list(lista_MaioresGastadores.index)
lista_Valores = list(lista_MaioresGastadores)
```

```python
sns.set(style="darkgrid",context="notebook", palette="RdBu")
_ = df.groupby("NomeEPartido").sum().sort_values(by="vlrliquido",ascending=False)["vlrliquido"].head(20).plot(kind="bar")
```

```python
df.sgpartido.unique()
```

```python
df.groupby("sgpartido").sum().sort_values(by="vlrliquido",ascending=False)["vlrliquido"].head(10).plot(kind="bar")
```

O PT aparece despontando no primeiro lugar em volume, porém isso deve ser devido ao maior número de parlamentares desse partido. Mais abaixo, a média foi calculada e o PT vai para o fim da lista de maiores gastdores.

```python
df.sgpartido.value_counts()
```

```python
qtd_partidos=len(df.sgpartido.unique())
print("Existem {} partidos.".format(qtd_partidos))
df_GastoMedioPorPartido = df.groupby("sgpartido").mean().sort_values(by="vlrliquido",ascending=False)["vlrliquido"].head(qtd_partidos)
df_GastoMedioPorPartido.plot(kind="bar");
```

```python
list(df_GastoMedioPorPartido.index[:5])
```

```python
list(df_GastoMedioPorPartido.index[-5:])
```

```python
df_MaioresGastadoresMedios = df[df["sgpartido"].isin(['AVANTE', 'PPL', 'PTB', 'PODE', 'PSC'])]
df_MenoresGastadoresMedios = df[df["sgpartido"].isin(['PV', 'REDE', 'NOVO', 'DC', 'LID.GOV-CD'])]
```

```python
df_MaioresGastadoresMedios
```

```python
_ = plt.subplot(2, 1, 1)
_ = sns.boxplot( x="sgpartido", y="vlrliquido", data=df_MaioresGastadoresMedios)
_ = plt.title("5 maiores gastadores", loc="left")
_ = plt.plot(figsize=(10,10))
_ = plt.ylabel("Valor Líquido da Nota")
_ = plt.xlabel("Partido")

_ = plt.subplot(2, 1, 2)
_ = sns.boxplot( x="sgpartido", y="vlrliquido", data=df_MenoresGastadoresMedios)
_ = plt.title("5 menores gastadores", loc="left")
_ = plt.plot(figsize=(10,10))
_ = plt.ylabel("Valor Líquido da Nota")
_ = plt.xlabel("Partido")

plt.show()
```
